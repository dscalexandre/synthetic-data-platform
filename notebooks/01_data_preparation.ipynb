### 1. IMPORTAÇÕES E CONFIGURAÇÃO

```python
# Importações

# Biblioteca padrão
import os

# Manipulação de dados
import pandas as pd

# Sistema de arquivos
from pathlib import Path

# Metadados do ambiente
%reload_ext watermark
%watermark --iversions
```

```python
# Configuração do diretório do projeto
project_root = Path.cwd()

if project_root.name == "notebooks":
    os.chdir(project_root.parent)

print(f"Diretório do projeto: {Path.cwd().name}")
```

```python
# Definir parâmetros de entrada e saída
input_directory = Path("data/raw/customers")
output_file = Path("data/processed/customers.csv")
file_extension = "*.csv"
encoding = "utf-8"
```

```python
# Localizar arquivos CSV no diretório de entrada
files = sorted(
    input_directory.glob(file_extension)
)

if not files:
    raise FileNotFoundError(f"Nenhum arquivo CSV encontrado em: {input_directory}")

print(f"Quantidade de arquivos encontrados: {len(files)}")

for file in files:
    print(file.name)
```

```python
# Carregar arquivos CSV em DataFrames
dataframes = [
    pd.read_csv(
        file,
        encoding=encoding
    )
    for file in files
]

print(f"Quantidade de DataFrames carregados: {len(dataframes)}")
```

```python
# Validar colunas dos datasets originais
reference_columns = list(dataframes[0].columns)

for file, dataframe in zip(files, dataframes):
    assert list(dataframe.columns) == reference_columns, (
        f"O arquivo {file.name} possui colunas diferentes da referência."
    )

print(f"Todos os arquivos possuem as mesmas {len(reference_columns)} colunas.")
```

```python
# Exibir quantidade total de registros e colunas dos datasets originais
total_records = sum(
    dataframe.shape[0]
    for dataframe in dataframes
)

total_columns = dataframes[0].shape[1]

print(f"Quantidade total de registros: {total_records:,}")
print(f"Quantidade total de colunas: {total_columns}")
```

```python
# Concatenar DataFrames em um único DataFrame
dataset = pd.concat(
    dataframes,
    ignore_index=True
)

print("Partições consolidadas com sucesso.")
```

```python
# Validar estrutura do DataFrame
print(f"Quantidade de registros: {dataset.shape[0]}")
print(f"Quantidade de colunas: {dataset.shape[1]}")
```

```python
# Visualizar as primeiras linhas do DataFrame
dataset.head()
```

```python
# Validar informações do DataFrame
dataset.info()
```

```python
# Verificar valores ausentes por coluna
dataset.isna().sum()
```

```python
# Verificar registros duplicados pelo identificador
duplicated_ids = dataset.duplicated(subset="id").sum()

print(f"Quantidade de IDs duplicados: {duplicated_ids}")
```

```python
# Exportar dataset consolidado
output_file.parent.mkdir(parents=True, exist_ok=True)

dataset.to_csv(
    output_file,
    index=False,
    encoding=encoding
)

print(f"Arquivo salvo em: {output_file}")
```

```python
# Confirmar exportação
file_exists = output_file.exists()

print(f"Arquivo criado: {file_exists}")
```
